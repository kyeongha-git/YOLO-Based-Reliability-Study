## Inference
- YOLO 모델로 타겟 부위만 추출하여 저장.
- Train / Valid / Test에 각각 진행하여 전체 데이터 모두에 적용.

```python
# Load the trained model
model = YOLO("/content/drive/MyDrive/weights_repair_replace/yolo_v8_repair_repalce_best.pt")

# Set the source directory and output directory for cropped images
source_dir = "/content/CarDD_repair_replace-6/valid/images"  # Update this path as needed
output_dir = "/content/cropped_images/valid"
os.makedirs(output_dir, exist_ok=True)

# Perform inference and crop bounding boxes
results = model.predict(source=source_dir, imgsz=416, save=False)

for result in results:
    image_path = result.path
    image = cv2.imread(image_path)

    for idx, box in enumerate(result.boxes.xyxy):  # Get bounding box coordinates
        x1, y1, x2, y2 = map(int, box)
        cropped_image = image[y1:y2, x1:x2]  # Crop the bounding box region

        # Save the cropped image
        base_name = os.path.basename(image_path)
        name, ext = os.path.splitext(base_name)
        cropped_path = os.path.join(output_dir, f"{name}_crop{idx}{ext}")
        cv2.imwrite(cropped_path, cropped_image)

print(f"Saved cropped images to {output_dir}")
```

```python
# 원본 경로와 클래스 목록
base_path = '/content/cropped_images/valid'
classes = ["repair", "replace"]

# 클래스 폴더 생성
for cls in classes:
    class_folder = os.path.join(base_path, cls)
    if not os.path.exists(class_folder):
        os.makedirs(class_folder)

# 파일 이동
for filename in os.listdir(base_path):
    file_path = os.path.join(base_path, filename)

    # 파일인지 확인
    if os.path.isfile(file_path):
        for cls in classes:
            if cls in filename:
                # 해당 클래스 폴더로 이동
                target_path = os.path.join(base_path, cls, filename)
                shutil.move(file_path, target_path)
                print(f"Moved: {filename} -> {cls}/")
                break
```
