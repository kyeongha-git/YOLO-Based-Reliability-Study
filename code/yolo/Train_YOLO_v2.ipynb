## Modify the yolov2.cfg file to fit the custom data

```python
cfg_file_path = 'cfg/yolov2.cfg'

def update_cfg_file(cfg_path, num_classes):
    with open(cfg_path, 'r') as f:
        cfg_lines = f.readlines()

    new_cfg_lines = []
    for i, line in enumerate(cfg_lines):
        # Set 'classes'
        if 'classes=' in line:
            cfg_lines[i] = f'classes={num_classes}\n'
        # Set 'filters' (convolutional layer just before the yolo layer)
        elif 'filters=' in cfg_lines[i-1]:
            filters_value = (num_classes + 5) * 5
            cfg_lines[i-1] = f'filters={filters_value}\n'

    # Save modified cfg file
    with open(cfg_path, 'w') as f:
        f.writelines(cfg_lines)

update_cfg_file(cfg_file_path, len(class_names))
```

## Move data (Download location -> train, valid and test / images)

```python
import shutil

# Move image files and corresponding label files within train and valid folders
def move_files(source_folder, target_folder, image_extensions):
    for file_name in os.listdir(source_folder):
        if any(file_name.endswith(ext) for ext in image_extensions):
            # Move image file
            shutil.move(os.path.join(source_folder, file_name), os.path.join(target_folder, file_name))

            # Move corresponding label file (.txt file)
            label_file_name = os.path.splitext(file_name)[0] + ".txt"
            if os.path.exists(os.path.join(source_folder, label_file_name)):
                shutil.move(os.path.join(source_folder, label_file_name), os.path.join(target_folder, label_file_name))

# List of image file extensions
image_extensions = ['.jpg', '.jpeg', '.png']

# Move image and label files to images folder
train_images_folder = os.path.join(dataset_location, 'train/images')
valid_images_folder = os.path.join(dataset_location, 'valid/images')
test_images_folder = os.path.join(dataset_location, 'test/images')

# Organize train and valid data
if not os.path.exists(train_images_folder):
    os.makedirs(train_images_folder)
if not os.path.exists(valid_images_folder):
    os.makedirs(valid_images_folder)
if not os.path.exists(test_images_folder):
    os.makedirs(test_images_folder)

move_files(train_images_path, train_images_folder, image_extensions)
move_files(valid_images_path, valid_images_folder, image_extensions)
move_files(test_images_path, test_images_folder, image_extensions)
```

## Predict data using trained weights and crop the bounding box to save it.
* Save only the bounding box with the highest confidence.
* If the prediction is below the threshold, save the original image in the NoDetection folder.

```python
# %cd ..
```

```python
import os
import cv2
import numpy as np

def crop_bounding_boxes(image_path, output_folder, net, output_layers, class_list, conf_threshold=0.1, nms_threshold=0.4):
    # Load the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not load image: {image_path}")
        return
    print(f"Loaded image size: {img.shape}")  # Debugging output

    height, width, _ = img.shape

    # YOLOv2 object detection
    blob = cv2.dnn.blobFromImage(img, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    layer_outputs = net.forward(output_layers)

    boxes = []
    confidences = []
    class_ids = []

    # Extract object information from each detection result
    for output in layer_outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > conf_threshold:
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                # Adjust to ensure bounding box does not exceed image bounds
                x = max(0, x)
                y = max(0, y)
                w = min(w, width - x)
                h = min(h, height - y)

                # Add only valid boxes with positive width and height
                if w > 0 and h > 0:
                    boxes.append([x, y, w, h])
                    confidences.append(float(confidence))
                    class_ids.append(class_id)

    # Handle case when no boxes are detected
    if len(boxes) == 0:
        print(f"No detected objects: {image_path}")

        # Create NoDetection folder if it does not exist
        no_detection_folder = os.path.join(output_folder, 'NoDetection')
        if not os.path.exists(no_detection_folder):
            os.makedirs(no_detection_folder)

        # Save the original image in the NoDetection folder
        save_path = os.path.join(no_detection_folder, os.path.basename(image_path))
        cv2.imwrite(save_path, img)
        print(f"Saved undetected image: {save_path}")
        return

    # Apply Non-Maximum Suppression (NMS)
    indices = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, nms_threshold)

    if len(indices) > 0:
        for i in indices.flatten():
            x, y, w, h = boxes[i]
            cropped_img = img[y:y+h, x:x+w]

            # Check if the cropped image is valid
            if cropped_img.size > 0 and w > 0 and h > 0:
                class_name = class_list[class_ids[i]]
                class_folder = os.path.join(output_folder, class_name)

                if not os.path.exists(class_folder):
                    os.makedirs(class_folder)

                save_path = os.path.join(class_folder, f"cropped_{os.path.basename(image_path)}_{i}.jpg")
                cv2.imwrite(save_path, cropped_img)
                print(f"Cropped and saved bounding box {i}: {save_path}")
            else:
                print(f"Invalid crop: {image_path}")
    else:
        print(f"No detected boxes: {image_path}")

# Function to load the model
def load_yolo_model(cfg_path, weights_path, names_path):
    net = cv2.dnn.readNet(weights_path, cfg_path)
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
    with open(names_path, "r") as f:
        classes = [line.strip() for line in f.readlines()]
    return net, output_layers, classes

# Set paths
cfg_path = "./darknet/cfg/yolov2.cfg"
weights_path = "./darknet/backup/yolov2_last.weights"
names_path = "./darknet/data/obj.names"

# Load YOLOv2 model
net, output_layers, class_list = load_yolo_model(cfg_path, weights_path, names_path)

# Input image folder and result save folder
input_folder = './additional_and_gan_yolov2-2/train/images'
output_folder = './additional_and_gan_yolov2-2/results/train'

# List of image files to process
image_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Execute detection and crop for each image
for image_path in image_files:
    crop_bounding_boxes(image_path, output_folder, net, output_layers, class_list)

print("Cropping completed for all images.")
```

## Count Your Folder

```python
def count_image_files_in_directory(directory_path):
    # Define a set of image file extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp'}

    try:
        # Check if the provided path exists and is a directory
        if not os.path.exists(directory_path):
            print(f"The path {directory_path} does not exist.")
            return
        if not os.path.isdir(directory_path):
            print(f"The path {directory_path} is not a directory.")
            return

        # Get the list of files in the directory
        files = os.listdir(directory_path)
        # Filter the files to include only images
        image_files = [file for file in files if os.path.splitext(file)[1].lower() in image_extensions]
        # Count the number of image files
        image_file_count = len(image_files)

        print(f"The number of image files in the directory '{directory_path.split('/')[-2] + '/' + directory_path.split('/')[-1]}' is: {image_file_count}")
    except Exception as e:
        print(f"An error occurred: {e}")
```

## Move it to the original class folder.
* If the predicted class does not match, Move it to the original class folder.

```python
import os
import shutil

# Set paths
train_folder = '/content/additional_and_gan_yolov2-2/results/test'

# List of class folders
class_folders = ['CombinationBreak', 'Others', 'Crack']

# Check images in the NoDetection folder and move them
no_detection_folder = os.path.join(train_folder, 'NoDetection')

# Check if the NoDetection folder exists
if os.path.exists(no_detection_folder):
    # Check the list of files in the folder
    for filename in os.listdir(no_detection_folder):
        if filename.endswith(('.jpg', '.jpeg', '.png')):  # Check image file extensions
            moved = False  # Initialize file move status
            print(f"Processing: {filename}")  # Print the name of the file being processed

            # Check if each class name is included in the file name and move it
            for class_folder in class_folders:
                # Compare ignoring case
                if class_folder.lower() in filename.lower():
                    # Set the path for the corresponding class folder
                    target_folder = os.path.join(train_folder, class_folder)

                    # Check if the class folder exists
                    if not os.path.exists(target_folder):
                        os.makedirs(target_folder)

                    # Move the file
                    source_file = os.path.join(no_detection_folder, filename)
                    target_file = os.path.join(target_folder, filename)
                    shutil.move(source_file, target_file)
                    print(f"{filename} has been moved to the {class_folder} folder.")
                    moved = True  # Indicate that the file has been moved
                    break  # Exit the loop since it only needs to be moved to one class

            if not moved:
                print(f"{filename} was not moved because there was no matching class folder.")
else:
    print("The NoDetection folder does not exist.")
```
